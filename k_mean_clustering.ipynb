```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/kkc2/Downloads\Mall_Customers.csv')

df.drop('Gender', axis=1, inplace=True)
```

```python
df.drop('CustomerID', axis=1, inplace=True)
```

```python
df
```

```python
wcss = []
for i in range(1, 21):
    km = KMeans(n_clusters = i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21), wcss)
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters = 8)
y_means = km.fit_predict(X)
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1], color='blue')
plt.scatter(X[y_means==1,0],X[y_means==1,1], color='red')
plt.scatter(X[y_means==2,0],X[y_means==2,1], color='green')
plt.scatter(X[y_means==3,0],X[y_means==3,1], color='yellow')
plt.scatter(X[y_means==4,0],X[y_means==4,1], color='purple')
plt.scatter(X[y_means==5,0],X[y_means==5,1], color='orange')
plt.scatter(X[y_means==6,0],X[y_means==6,1], color='black')
plt.scatter(X[y_means==7,0],X[y_means==7,1], color='voilet')
```
